## Interview Questions

## 1.	What is lift and why is it important in Association rules?

Lift is a metric used in association rule mining to evaluate the strength of a rule by comparing the observed frequency of the rule to the expected frequency if the items were independent. It helps determine whether the presence of one item (or itemset) in a transaction increases the likelihood of another item appearing.

## 2.	What is support and Confidence. How do you calculate them?

Support

Definition: Support measures the proportion of transactions in a dataset that contain a specific itemset (or rule). It indicates how frequently the itemset appears in the dataset

Calculation:

Support(A)=Total number of transactions / Number of transactions containing A

Confidence

Definition: Confidence measures the likelihood that a rule is true. Specifically, it calculates the probability that an item B
B is purchased given that item A is purchased. It indicates the strength of the implication in the rule 𝐴→𝐵.

Calculation:

Confidence(A→B)= Number of transactions containing A / Number of transactions containing both A and B
