2.Statastical Summary

Mean=The mean (or average) is the sum of all values divided by the number of values. It provides a measure of the central location of the data.            
Median=The median is the middle value when the data is sorted in ascending order.It is a measure of central tendency that is less sensitive to outliers than the mean.
Interquartile Range (IQR)=The IQR measures the range within which the middle 50% of the data falls. It is the difference between the 75th percentile (Q3) and the 25th percentile (Q1). It helps to understand the spread of the central portion of the data and is useful for identifying outliers.

```python
def remove_outlier(col):
    sorted(col)
    q1,q3=col.quantile([0.25,0.75])
    IQR=q3-q1
    lower_extreme=q1-(1.5*IQR)
    upper_extreme=q3+(1.5*IQR)
    return lower_extreme,upper_extreme
```

```python
low,high=remove_outlier(df['MSTV'])
low,high
```

```python
df['MSTV']=np.where(df['MSTV']>high,low,df['MSTV'])
df['MSTV']=np.where(df['MSTV']<low,high,df['MSTV'])
```

```python
df.boxplot(column='MSTV')
plt.show()
```

```python
def correlation(dataset,thershold):
    corr= dataset.corr()
    col_corr=set()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j])>thershold:
                col_name= corr.columns[i]
                col_corr.add(col_name)
    return col_corr
```
